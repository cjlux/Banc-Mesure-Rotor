Syntaxe : `plot_ROTOR(files[n], xyz=(0/1, 0/1, 0/1), figsize=(width, height));`

- `figsize=(width, height)` : optional, to force width and height (inch) of the figure.

## 3 $-$ Power Spectral Density (PSD) Plot<a name="3"></a>

Syntaxe : `plot_ROTOR(files[n], xyz=(0/1, 0/1, 0/1), fft=True, figsize=(width, height));`
- `n` : file number
- `xyz` : 101, 111... which component X,Y,Z to plot
- `fft=True` : to compute the FFT of X,Y,Z
- `figsize=(width, height)` : optional, to force width and height (inch) of the figure.

```python
 plot_ROTOR(files[11], xyz=(1,0,1), fft=True);
```
